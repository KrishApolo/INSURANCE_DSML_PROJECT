###**Confidence intervals**
The formula for a confidence interval for a population mean is:CI = x  t*(s/n)where x is the sample mean, s is the sample standard deviation, n is the sample size, and t is the t-value from the t-distribution with n-1 degrees of freedom and the desired level of confidence.

Confidence interval analysis of average premium price for given sample data:

```python
df.PremiumPrice.describe()
```

```python
import scipy.stats
from scipy.stats import kruskal,shapiro,boxcox,ttest_ind,f_oneway,levene,chisquare,chi2,chi2_contingency
```

```python
sample_mean=df.PremiumPrice.mean()
std_err=df.PremiumPrice.std()
t=scipy.stats.t.ppf(q=0.95,df=(df.PremiumPrice.count()-1))
lower_limit = sample_mean-(t*std_err/(np.sqrt(df.PremiumPrice.count())))
upper_limit = sample_mean+(t*std_err/(np.sqrt(df.PremiumPrice.count())))

t95=scipy.stats.t.ppf(q=0.975,df=(df.PremiumPrice.count()-1))
lower_limit95 = sample_mean-(t95*std_err/(np.sqrt(df.PremiumPrice.count())))
upper_limit95 = sample_mean+(t95*std_err/(np.sqrt(df.PremiumPrice.count())))

t99=scipy.stats.t.ppf(q=0.995,df=(df.PremiumPrice.count()-1))
lower_limit99 = sample_mean-(t99*std_err/(np.sqrt(df.PremiumPrice.count())))
upper_limit99 = sample_mean+(t99*std_err/(np.sqrt(df.PremiumPrice.count())))
```

```python
print("90% CI of Premium Price is",lower_limit,' to ',upper_limit)
print("95% CI of Premium Price is",lower_limit95,' to ',upper_limit95)
print("99% CI of Premium Price is",lower_limit99,' to ',upper_limit99)
```

###**Relation between Independent factors and Premium Price**

H0 = Diabetes and PremiumPrice are independent

HA = Diabetes and PremiumPrice are dependent

```python
df_d0=df[df['Diabetes']==0]
df_d1=df[df['Diabetes']==1]
```

```python
stats,p=ttest_ind(df_d0['PremiumPrice'],df_d1['PremiumPrice'])
alpha=0.05
print(stats,p)
if (p<alpha):
  print('reject H0')
else:
  print('failed to reject H0')
```

###**Average PremiumPrice is similar for different number of major surgeries**

H0 = PremiumPrice is independent of number of major surgeries

Ha = PremiumPrice is dependent of number of major surgeries

```python
df.NumberOfMajorSurgeries.unique()
```

```python
df_s0=df[df['NumberOfMajorSurgeries']==0]
df_s1=df[df['NumberOfMajorSurgeries']==1]
df_s2=df[df['NumberOfMajorSurgeries']==2]
df_s3=df[df['NumberOfMajorSurgeries']==3]
```

```python
plt.hist(df_s1['PremiumPrice'])
```

sample distributions are not guassian

```python
df_s0mean=[df_s0['PremiumPrice'].sample(3000,replace= True).mean() for i in range(3000)]
df_s1mean=[df_s1['PremiumPrice'].sample(3000,replace= True).mean() for i in range(3000)]
df_s2mean=[df_s2['PremiumPrice'].sample(3000,replace= True).mean() for i in range(3000)]
df_s3mean=[df_s3['PremiumPrice'].sample(3000,replace= True).mean() for i in range(3000)]
```

```python
plt.hist(df_s1mean)
```

```python
dfs0=boxcox(df_s0mean,lmbda=0.27)
dfs1=boxcox(df_s1mean,lmbda=0.27)
dfs2=boxcox(df_s2mean,lmbda=0.27)
dfs3=boxcox(df_s3mean,lmbda=0.27)
plt.hist(dfs1)
```

```python
stats,p=shapiro(dfs1)
alpha=0.05
print(stats,p)
print('shapiro H0 is distribution is guassian')
if (p<alpha):
  print('reject H0')
else:
  print('failed to reject H0')
```

```python
stats, p = levene(dfs1,dfs2,dfs3,dfs0,center='mean')
alpha=0.01
print(stats,p)
print('levenes test h0 is groups have equal variances')
if (p<alpha):
  print('reject H0')
else:
  print('failed to reject H0')
```

Since groups have no equal variances we use kruskal test to test the dependency

```python
stats, p = kruskal(dfs0,dfs1,dfs2,dfs3)
alpha=0.05
print(stats,p)
print('kruskuls test h0 is groups are independent')
if (p<alpha):
  print('reject H0')
else:
  print('failed to reject H0')
```

Hence from the insighst available we conclude that premium price is dependent on number of major surgeries.
